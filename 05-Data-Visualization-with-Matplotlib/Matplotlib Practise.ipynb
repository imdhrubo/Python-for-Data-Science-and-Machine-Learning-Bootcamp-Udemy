```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x= np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
# functional plot
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
# plt.show() use this if not in notebook
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented plot
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Bigger plot')

axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

for current_axe in axes:
    current_axe.plot(x,y)

plt.tight_layout()
#axes.plot(x,y,'g')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(12,3))

axes[0].plot(x, y, 'r')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title1')

axes[1].plot(y, x, 'r')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('title2')

plt.tight_layout()
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax1.set_title('title')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_title('title')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z,'b')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_xlim([0, 100])
ax1.set_ylim([0,10000])


ax2.plot(x,y,'b')
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig = plt.subplots(nrows=1,ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,color='blue',lw=3,ls='--')
axes[1].plot(x,z,color='red',lw=3)

fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12, 2))

axes[0].plot(x,y,color='blue',lw=3)
axes[1].plot(x,z,color='red',lw=3, ls='--')

fig.tight_layout()

```
